# Data Pre-Processing 

```python
# Subscription_Start_Date         object
# Subscription_End_Date           object

df['Subscription_Start_Date'] = pd.to_datetime(df['Subscription_Start_Date'])
df['Subscription_End_Date'] = pd.to_datetime(df['Subscription_End_Date'])
```

```python
df.dtypes
```

```python
df.isna().sum()
```

```python
df.loc[df.duplicated()]
```

```python
#Handling Missing Values

#Impute Monthly cost based on subscription tier median

#Calculate the median monthly cost for each Subscription_Tier
tiers_median = df.groupby('Subscription_Tier')['Monthly_Cost'].median()

#Impute missing values using the tier-specific medians
df['Monthly_Cost'] = df.apply(
    lambda row : tiers_median[row['Subscription_Tier']] if pd.isna(row['Monthly_Cost']) else row['Monthly_Cost'], axis=1
)   #axis=1 is row wise
```

```python
df['Monthly_Cost'].isna().sum()
```

```python
#Data Consistency

#Group by User_ID and calculate the sum of Monthly_Cost and Annual_Cost
user_total_costs = df.groupby("User_ID").agg({
    "Monthly_Cost": "sum",  # Sum up monthly costs for each user
    "Annual_Cost": "sum"    # Sum up annual costs for each user
}).reset_index()

#  Rename columns for clarity
user_total_costs = user_total_costs.rename(columns={
    "Monthly_Cost": "Recalculated_Total_Spend_Monthly",
    "Annual_Cost": "Recalculated_Total_Spend_Annual"
})

# Merge recalculated totals back into the main DataFrame
df = pd.merge(df, user_total_costs, on="User_ID", how="left")

# Update the existing Total_Spend columns with recalculated values
df['Total_Spend_Monthly'] = df['Recalculated_Total_Spend_Monthly']
df['Total_Spend_Annual'] = df['Recalculated_Total_Spend_Annual']

#Drop the temporary recalculated columns
df.drop(['Recalculated_Total_Spend_Monthly', 'Recalculated_Total_Spend_Annual'], axis=1, inplace=True)
```

```python
df['Frequency'] =df.apply(
    lambda row: 'Monthly' if pd.isna(row['Frequency']) and pd.notnull(row['Monthly_Cost'])  
                else ('Annual' if pd.isna(row['Frequency']) and row['Annual_Cost']>0 else row['Frequency']) , axis =1
)
```

```python
df['Frequency'].isna().sum()
```

```python
df['Redundant_Service'].fillna("No Redundant Service", inplace =True) 
```

```python
df['Redundant_Service'].isna().sum()
```

```python
df['Cheaper_Alternative'].fillna("None", inplace =True)
df['Cheaper_Alternative'].isna().sum()
```

```python
df['Subscription_End_Date'] = df['Subscription_End_Date'].fillna(pd.NaT)
```

```python
df['Cancellation_Reason'] =df.apply(
    lambda row: 'Active Subscription' if pd.isna(row['Subscription_End_Date']) 
    else ('Unknown' if pd.isna(row['Cancellation_Reason']) else row['Cancellation_Reason']) , axis = 1
)
```

```python
df['Subscription_Duration_Days'] = df.apply(
    lambda row: (row['Subscription_End_Date'] - row['Subscription_Start_Date']).days 
    if (pd.notna(row['Subscription_Start_Date']) and pd.notna(row['Subscription_End_Date']))
    else ((datetime.today() - row['Subscription_Start_Date']).days 
    if pd.notna(row['Subscription_Start_Date'])
    else None), 
    axis = 1)
```
